## 2. NumPy – Numerical Computing

````python
import numpy as np

print(f"NumPy version: {np.__version__}")

# --- Arrays ---
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("Array a:", a)
print("Array b:", b)
print("Element-wise multiplication:", a * b)
print("Dot product:", np.dot(a, b))

# --- 2D Array (Matrix) ---
matrix = np.arange(1, 10).reshape(3, 3)
print("\n3x3 Matrix:\n", matrix)
print("Transpose:\n", matrix.T)
print("Matrix determinant:", round(np.linalg.det(matrix), 2))

# --- Statistical operations ---
data = np.random.randn(1000)  # 1000 random samples from standard normal
print(f"\nMean: {data.mean():.4f}, Std: {data.std():.4f}, Min: {data.min():.4f}, Max: {data.max():.4f}")
````

## 3. Pandas – Data Manipulation

````python
import pandas as pd

print(f"Pandas version: {pd.__version__}")

# --- Load Iris CSV from scikit-learn and convert to DataFrame ---
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print("\nFirst 5 rows:")
print(df.head())

print("\nDataset info:")
df.info()

print("\nDescriptive statistics:")
print(df.describe())

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nValue counts for 'species':")
print(df['species'].value_counts())
````

## 4. Matplotlib – Visualization

````python
import matplotlib
import matplotlib.pyplot as plt

print(f"Matplotlib version: {matplotlib.__version__}")

# Scatter plot: sepal length vs sepal width coloured by species
colours = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for species, colour in colours.items():
    subset = df[df['species'] == species]
    axes[0].scatter(subset['sepal length (cm)'], subset['sepal width (cm)'],
                    c=colour, label=species, alpha=0.7)
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Sepal Width (cm)')
axes[0].set_title('Iris: Sepal Length vs Width')
axes[0].legend()

# Histogram of petal length
axes[1].hist(df['petal length (cm)'], bins=20, color='steelblue', edgecolor='black')
axes[1].set_xlabel('Petal Length (cm)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Petal Length Distribution')

plt.tight_layout()
plt.show()
````

## 5. Seaborn – Statistical Visualization

````python
import seaborn as sns

print(f"Seaborn version: {sns.__version__}")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Pairplot (using seaborn's built-in iris)
sns.boxplot(data=df, x='species', y='petal length (cm)', palette='Set2', ax=axes[0])
axes[0].set_title('Petal Length by Species')

# Correlation heatmap
corr = df.drop(columns=['target', 'species']).corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[1])
axes[1].set_title('Feature Correlation Heatmap')

plt.tight_layout()
plt.show()
````

## 6. scikit-learn – ML Utilities

````python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

print(f"scikit-learn version: {sklearn.__version__}")

X = df.drop(columns=['target', 'species'])
y = df['target']  # 0, 1, 2

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# Build a pipeline: scaler + classifier
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=200))
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

print(f"\nTest Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
````

## 7. SciPy – Scientific Computing

````python
import scipy
from scipy import stats

print(f"SciPy version: {scipy.__version__}")

# T-test: is the mean sepal length of setosa different from versicolor?
setosa_sl = df[df['species'] == 'setosa']['sepal length (cm)']
versicolor_sl = df[df['species'] == 'versicolor']['sepal length (cm)']

t_stat, p_value = stats.ttest_ind(setosa_sl, versicolor_sl)
print(f"\nTwo-sample t-test (setosa vs versicolor sepal length):")
print(f"  t-statistic = {t_stat:.4f}")
print(f"  p-value     = {p_value:.6f}")
print(f"  Conclusion  : {'Significant difference' if p_value < 0.05 else 'No significant difference'}")
````

## 8. TensorFlow / Keras – Deep Learning (Google)

````python
try:
    import tensorflow as tf
    from tensorflow import keras

    print(f"TensorFlow version: {tf.__version__}")
    print(f"Keras version: {keras.__version__}")

    # Quick demo: build and summarise a small MLP
    model = keras.Sequential([
        keras.layers.Dense(16, activation="relu", input_shape=(4,)),
        keras.layers.Dense(8,  activation="relu"),
        keras.layers.Dense(3,  activation="softmax")   # 3 iris classes
    ])
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    model.summary()
except ImportError:
    print("TensorFlow not installed. Run: pip install tensorflow")
````

## 9. PyTorch – Deep Learning (Meta)

````python
try:
    import torch
    import torch.nn as nn

    print(f"PyTorch version: {torch.__version__}")
    print(f"CUDA available : {torch.cuda.is_available()}")

    # Quick demo: build a small MLP using PyTorch
    class SimpleMLP(nn.Module):
        def __init__(self):
            super().__init__()
            self.net = nn.Sequential(
                nn.Linear(4, 16),
                nn.ReLU(),
                nn.Linear(16, 8),
                nn.ReLU(),
                nn.Linear(8, 3)     # 3 classes
            )

        def forward(self, x):
            return self.net(x)

    pytorch_model = SimpleMLP()
    print(pytorch_model)
    print(f"Total parameters: {sum(p.numel() for p in pytorch_model.parameters())}")
except ImportError:
    print("PyTorch not installed. Run: pip install torch")
````

## 10. XGBoost – Gradient Boosting

````python
try:
    import xgboost as xgb

    print(f"XGBoost version: {xgb.__version__}")

    xgb_clf = xgb.XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric="mlogloss", random_state=42)
    xgb_clf.fit(X_train, y_train)
    xgb_pred = xgb_clf.predict(X_test)
    print(f"XGBoost Test Accuracy: {accuracy_score(y_test, xgb_pred):.4f}")
except ImportError:
    print("XGBoost not installed. Run: pip install xgboost")
````
